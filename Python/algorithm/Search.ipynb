#  Binary Search

```python
def binary_search(nums: List[int], target: int) -> int:
    left = 0
    right = len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid -1
    return -1   
```

```python
nums = [1,2,2,2,3]
binary_search(nums, 2)
```

```python
def binary_search_left(nums: List[int], target: int) -> int:
    left = 0
    right = len(nums)
    
    while left < right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            right = mid
        elif nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid
    return left 
```

```python
binary_search_left(nums, 2)
```

```python
def binary_search_right(nums: List[int], target: int) -> int:
    left = 0
    right = len(nums)
    
    while left < right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            left = mid + 1
        elif nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid
    return left - 1
```

```python
binary_search_right(nums, 2)
```

# Koko Balana


```python
from math import  ceil
def canFinish(piles: List[int], speed: int, H: int):
    return sum([timeof(n, speed) for n in piles]) <= H

def timeof(n: int, speed: int):
    return ceil(n / speed)
```

```python
def min_eating_speed_forch(piles: List[int], H: int):
    _max = max(piles)
    for speed in range(1, _max):
        if (canFinish(piles, speed, H)):
            return speed
    
    return _max
```

```python
piles = [3, 6,  7, 11]
H = 8
min_eating_speed_forch(piles, H)
```

```python
piles = [30, 11, 23, 4, 20]
H = 5
min_eating_speed_forch(piles, H)
```

```python
def min_eating_speed_bs(piles: List[int], H: int):
    left = 1
    right = max(piles) + 1
    
    while left < right:
        mid = left + (right - left) // 2
        if canFinish(piles, mid, H):
            right = mid
        else:
            left = mid + 1
    return left
```

```python
piles = [30, 11, 23, 4, 20]
H = 5
min_eating_speed_bs(piles, H)
```
